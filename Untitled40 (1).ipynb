Model Implementation

```python
import numpy as np

class Dynamic_parkingpricer:
    """
    Dynamic Parking Pricing System
    -------------------------------------------------
    Description:
    This system implements 3 pricing models to simulate
    real-world smart urban parking pricing strategies.

    Models:
      - Model 1: Linear Pricing based on occupancy
      - Model 2: Demand-Based Pricing (occupancy, traffic, etc.)
      - Model 3: Competitive Pricing based on nearby lots
    """

    def __init__(self, base_price=10.0, min_price=5.0, max_price=20.0):
        # Base price and boundary limits
        self.base_price = base_price
        self.min_price = min_price
        self.max_price = max_price

    # ------------------------------------------------
    # VEHICLE WEIGHT LOGIC
    # ------------------------------------------------
    def get_vehicle_weights(self, vehicle_type, occupancy_rate, hour, is_special_day):
        """
        Applies weight multipliers based on vehicle type and conditions.
        Bigger vehicles = more charge; lighter = less.
        """
        base_weights = {
            'car': 1.0,
            'truck': 1.4,
            'bike': 0.6,
            'cycle': 0.4
        }

        weight = base_weights.get(vehicle_type, 1.0)

        # Discount for bike/cycle during low usage
        if occupancy_rate < 0.4 and vehicle_type in ['bike', 'cycle']:
            weight *= 0.9

        # Penalty for trucks during rush hour
        if hour in [11, 12, 13, 14] and vehicle_type == 'truck':
            weight *= 1.2

        # Special day slight increase for cars/trucks
        if is_special_day and vehicle_type in ['car', 'truck']:
            weight *= 1.1

        return weight

    # ------------------------------------------------
    # MODEL 1: LINEAR PRICING
    # ------------------------------------------------
    def model1_linear_pricing(self, prev_price, occupancy, capacity, alpha=5.0):
        """
        Price = Previous_Price +  * (Occupancy / Capacity)
        """
        occupancy_rate = occupancy / capacity if capacity > 0 else 0
        new_price = prev_price + alpha * occupancy_rate
        return np.clip(new_price, self.min_price, self.max_price)

    # ------------------------------------------------
    # MODEL 2: DEMAND-BASED PRICING
    # ------------------------------------------------
    def model2_demand_based_pricing(self, occupancy, capacity, queue_length,
                                    traffic_level, is_special_day, vehicle_type, hour):
        """
        Demand = (Occupancy/Capacity) + Queue  Traffic + IsSpecialDay + VehicleWeight
        Price = BasePrice  (1 + tanh(Demand/10))  VehicleWeight
        """
        occupancy_rate = occupancy / capacity if capacity > 0 else 0
        normalized_queue = min(queue_length / 20.0, 1.0)
        vehicle_weight = self.get_vehicle_weights(vehicle_type, occupancy_rate, hour, is_special_day)

        traffic_penalty = {
            'low': -0.5,
            'average': 0.0,
            'high': 0.5
        }.get(traffic_level, 0.0)

        special_day_value = 1 if is_special_day else 0

        # Model coefficients (customized)
        alpha = 8.0      # Occupancy impact
        beta = 3.0       # Queue length impact
        gamma = 2.0      # Traffic level impact
        delta = -1.0     #  Special day = slight discount (was +1.5 before)
        epsilon = 1.0    # Vehicle type impact
        lambd = 0.5      # Price responsiveness

        # Demand calculation
        demand = (
            alpha * occupancy_rate +
            beta * normalized_queue -
            gamma * traffic_penalty +
            delta * special_day_value +
            epsilon * (vehicle_weight - 1.0)
        )

        normalized_demand = np.tanh(demand / 10.0)
        price = self.base_price * (1 + lambd * normalized_demand)
        price *= vehicle_weight

        return np.clip(price, self.min_price, self.max_price), demand

    # ------------------------------------------------
    # HAVERSINE DISTANCE FUNCTION
    # ------------------------------------------------
    def calculate_distance(self, lat1, lon1, lat2, lon2):
        """
        Returns distance (in km) between two GPS points using the Haversine formula.
        """
        R = 6371  # Earth radius in km
        dlat = np.radians(lat2 - lat1)
        dlon = np.radians(lon2 - lon1)

        a = np.sin(dlat / 2) ** 2 + \
            np.cos(np.radians(lat1)) * np.cos(np.radians(lat2)) * \
            np.sin(dlon / 2) ** 2

        c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
        return R * c

    # ------------------------------------------------
    # MODEL 3: COMPETITIVE PRICING MODEL
    # ------------------------------------------------
    def model3_competitive_pricing(self, current_lot, all_lots_df, current_price, proximity_km=2.0):
        """
        Adjusts current lot price based on nearby competitor occupancy and proximity.
        Suggests alternate lots if better.
        """
        lat_c, lon_c = current_lot['Latitude'], current_lot['Longitude']
        occ_c, cap_c = current_lot['Occupancy'], current_lot['Capacity']
        occ_rate_c = occ_c / cap_c if cap_c > 0 else 0.0

        competitors = []

        for _, lot in all_lots_df.iterrows():
            if lot['SystemCodeNumber'] == current_lot['SystemCodeNumber']:
                continue  # skip self

            dist = self.calculate_distance(lat_c, lon_c, lot['Latitude'], lot['Longitude'])
            if dist <= proximity_km:
                occ_rate = lot['Occupancy'] / lot['Capacity'] if lot['Capacity'] > 0 else 0
                competitors.append({
                    'code': lot['SystemCodeNumber'],
                    'distance': dist,
                    'occ_rate': occ_rate
                })

        if not competitors:
            return current_price, None

        avg_occ = np.mean([c['occ_rate'] for c in competitors])

        if occ_rate_c > 0.9:
            if avg_occ < 0.7:
                best = min(competitors, key=lambda x: x['occ_rate'])
                suggestion = f" Suggest lot {best['code']} ({best['occ_rate']*100:.1f}% full, {best['distance']:.1f} km away)"
                new_price = current_price * 0.93
            else:
                suggestion = None
                new_price = current_price * 1.1
        elif occ_rate_c < 0.6:
            if avg_occ > 0.8:
                new_price = current_price * 1.05
            else:
                new_price = current_price * 0.97
            suggestion = None
        else:
            new_price = current_price
            suggestion = None

        return np.clip(new_price, self.min_price, self.max_price), suggestion
```

```python
def apply_all_models_on_data(dataframe):
    """
    Applies all three pricing models to the provided parking dataset.
    Returns a DataFrame with computed prices and demand scores.
    """

    pricing_output = []

    # Loop over each unique parking lot
    for lot_id in dataframe['SystemCodeNumber'].unique():
        lot_data = dataframe[dataframe['SystemCodeNumber'] == lot_id].copy()
        lot_data = lot_data.sort_values('Timestamp')  # Ensure chronological order

        previous_price = pricer.base_price  # Start with the default base price

        for _, row in lot_data.iterrows():
            # ---- MODEL 1 ----
            linear_price = pricer.model1_linear_pricing(
                previous_price,
                row['Occupancy'],
                row['Capacity']
            )

            # ---- MODEL 2 ----
            demand_price, demand_value = pricer.model2_demand_based_pricing(
                occupancy=row['Occupancy'],
                capacity=row['Capacity'],
                queue_length=row['QueueLength'],
                traffic_level=row['TrafficConditionNearby'],
                is_special_day=row['IsSpecialDay'],
                vehicle_type=row['VehicleType'],
                hour=row['Hour']
            )

            # ---- MODEL 3 ----
            competitive_price, suggestion = pricer.model3_competitive_pricing(
                current_lot=row,
                all_lots_df=dataframe,
                current_price=demand_price
            )

            pricing_output.append({
                'Timestamp': row['Timestamp'],
                'SystemCodeNumber': row['SystemCodeNumber'],
                'Capacity': row['Capacity'],
                'Occupancy': row['Occupancy'],
                'OccupancyRate': row['Occupancy'] / row['Capacity'],
                'QueueLength': row['QueueLength'],
                'Traffic': row['TrafficConditionNearby'],
                'IsSpecialDay': row['IsSpecialDay'],
                'VehicleType': row['VehicleType'],
                'Hour': row['Hour'],
                'LinearPrice_Model1': linear_price,
                'DemandPrice_Model2': demand_price,
                'CompetitivePrice_Model3': competitive_price,
                'DemandScore': demand_value,
                'RoutingSuggestion': suggestion
            })

            previous_price = demand_price

    return pd.DataFrame(pricing_output)
```

```python
import time
from IPython.display import clear_output

def simulate_real_time(df_stream, delay=1):
    """
    Simulates streaming of parking data row-by-row
    """
    df_stream = df_stream.sort_values('Timestamp')
    prev_price = pricer.base_price

    for idx, row in df_stream.iterrows():
        # Apply all 3 models
        price_m1 = pricer.model1_linear_pricing(prev_price, row['Occupancy'], row['Capacity'])
        price_m2, demand_score = pricer.model2_demand_based_pricing(
            row['Occupancy'], row['Capacity'], row['QueueLength'],
            row['TrafficConditionNearby'], row['IsSpecialDay'], row['VehicleType'], row['Hour']
        )
        price_m3, suggestion = pricer.model3_competitive_pricing(row, df_stream, price_m2)

        # Print simulated output
        clear_output(wait=True)
        print(" Real-Time Data Received:")
        print(f"Timestamp: {row['Timestamp']}")  #  changed from 'DateTime'
        print(f"SystemCode: {row['SystemCodeNumber']}")
        print(f"Occupancy: {row['Occupancy']} / {row['Capacity']}")
        print(f"VehicleType: {row['VehicleType']} | Traffic: {row['TrafficConditionNearby']}")
        print(f"Special Day: {row['IsSpecialDay']}")
        print(f"\n Model 1 Price: {price_m1:.2f}")
        print(f" Model 2 Price: {price_m2:.2f} (Demand Score: {demand_score:.2f})")
        print(f" Model 3 Price: {price_m3:.2f}")
        if suggestion:
            print(f"Suggestion: {suggestion}")

        prev_price = price_m2  # update for next iteration
        time.sleep(delay)  # Simulate delay

# Run simulation on a small sample
simulate_real_time(df_sample.head(10), delay=2)
```

**Pathway Streaming Simulation**

```python
# Pathway Streaming Simulation (Model 2 Only)

class ParkingDataStream:
    def __init__(self, df):
        self.df = df.copy()
        self.df = self.df.sort_values('Timestamp')

    def create_pathway_stream(self):
        stream_data = self.df.head(100).copy()
        return pw.debug.table_from_pandas(stream_data)

stream_simulator = ParkingDataStream(df_sample)
input_stream = stream_simulator.create_pathway_stream()

@pw.udf
def demand_based_price_udf(occupancy, capacity, queue_length, traffic, is_special_day, vehicle_type, hour):
    price, demand = pricer.model2_demand_based_pricing(
        occupancy, capacity, queue_length, traffic, is_special_day, vehicle_type, hour
    )
    return price, demand

processed = input_stream.select(
    **input_stream,
    price_and_demand = demand_based_price_udf(
        pw.this.Occupancy, pw.this.Capacity, pw.this.QueueLength,
        pw.this.TrafficConditionNearby, pw.this.IsSpecialDay,
        pw.this.VehicleType, pw.this.Hour
    )
)

final_stream = processed.select(
    **processed,
    Price = processed.price_and_demand[0],
    DemandScore = processed.price_and_demand[1],
    OccupancyRate = pw.this.Occupancy / pw.this.Capacity
)

pw.io.jsonlines.write(final_stream, "./streaming_output.jsonl")
pw.run()
```
